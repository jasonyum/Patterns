# rand7 and rand5

- rand7() creates 1 to 7 integer (inclusive) uniformly
- rand5() creates 1 to 5 inter (inclusive) uniformly 

It's kind of weird that they're asking for two functions that are doing some pretty similar things. Obviously we could do this really simply if we imported numpy lol. 

```python
# Obviously not what they're asking
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def rand7Test(): 
    chosen = np.random.uniform(0.5, 7.5) # excludes 8
    return round(chosen) 
```

```python
lst = [rand7Test() for i in range(10000)]
```

```python
plt.hist(lst, bins='auto', rwidth=1)
```

```python
sns.distplot(lst)
```
